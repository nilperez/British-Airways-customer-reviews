## Split and train the data
###Dividir los datos y entrenar
Splitting the dataset into training and testing sets using the train_test_split function from the sklearn.model_selection module.

Dividimos el dataset en conjuntos de entrenamiento y prueba utilizando la función train_test_split del módulo sklearn.model_selection.





```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/BA/customer_booking_clean.csv", encoding="ISO-8859-1")
df.head()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(columns=['booking_complete'], axis=1)
y = df['booking_complete']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

The resulting variables x_train, x_test, y_train, and y_test contain the training and testing sets of the features and target variable, respectively.

Las variables resultantes x_train, x_test, y_train y y_test contienen los conjuntos de entrenamiento y prueba de las características y la variable objetivo, respectivamente.

```python
y_train.value_counts(normalize=True)
```

The result above is the class distribution of the target variable booking_complete after splitting the data into training and testing sets.

The values 0.851136 and 0.148864 represent the proportion of the samples that belong to each class in the training set. In this case, the class distribution is unbalanced, with the majority of the samples (85.11%) belonging to one class and the minority of the samples (14.89%) belonging to the other class.

El resultado anterior es la distribución de clases de la variable objetivo "booking_complete" después de dividir los datos en conjuntos de entrenamiento y prueba.

Los valores 0.851136 y 0.148864 representan la proporción de las muestras que pertenecen a cada clase en el conjunto de entrenamiento. En este caso, la distribución de clases está desbalanceada, con la mayoría de las muestras (85.11%) perteneciendo a una clase y la minoría de las muestras (14.89%) perteneciendo a la otra clase.

```python
# Importar la función de remuestreo
from sklearn.utils import resample

# Separar el conjunto de datos en clases mayoritarias y minoritarias
df_majority = df[df.booking_complete == 0]  # Clase mayoritaria
df_minority = df[df.booking_complete == 1]  # Clase minoritaria

# Upsample de la clase minoritaria
df_minority_upsampled = resample(df_minority,
                                 replace=True,     # muestreo con reemplazo
                                 n_samples=len(df_majority),    # para igualar la clase mayoritaria
                                 random_state=42)  # resultados reproducibles

# Combinar la clase mayoritaria con la clase minoritaria aumentada
df_upsampled = pd.concat([df_majority, df_minority_upsampled])

# Verificar la distribución de clases
print(df_upsampled.booking_complete.value_counts(normalize=True))
```
